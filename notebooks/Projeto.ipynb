```python
#CONTRIBUIÇÃO DO COLEGA RODRIGO

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Carregar a base
df = pd.read_csv("acidentes2025.csv", encoding="ISO-8859-1", sep=";")

# Contar acidentes por UF
acidentes_por_uf = df["uf"].value_counts().reset_index()
acidentes_por_uf.columns = ["UF", "Total_Acidentes"]

# Dicionário de regiões por estado
regioes = {
    "Norte": ["AC","AP","AM","PA","RO","RR","TO"],
    "Nordeste": ["AL","BA","CE","MA","PB","PE","PI","RN","SE"],
    "Centro-Oeste": ["GO","MT","MS","DF"],
    "Sudeste": ["ES","MG","RJ","SP"],
    "Sul": ["PR","RS","SC"]
}

# Criar coluna 'Região' no DataFrame
def identificar_regiao(uf):
    for regiao, estados in regioes.items():
        if uf in estados:
            return regiao
    return "Desconhecido"

acidentes_por_uf["Região"] = acidentes_por_uf["UF"].apply(identificar_regiao)

# Configurar estilo
sns.set_theme(style="whitegrid")

# Criar gráfico de barras colorido por região
plt.figure(figsize=(12, 6))
grafico = sns.barplot(
    data=acidentes_por_uf,
    x="UF",
    y="Total_Acidentes",
    hue="Região",          # cores por região
    dodge=False,           # barras não duplicadas
    palette="Set2"         # paleta de cores agradável
)

# Adicionar rótulos acima das barras
for p in grafico.patches:
    grafico.annotate(
        format(p.get_height(), ".0f"),
        (p.get_x() + p.get_width() / 2., p.get_height()),
        ha="center", va="bottom",
        color="black", weight="bold", fontsize=9
    )

# Personalizar layout
plt.title("Número de acidentes por UF (2025)", fontsize=18, weight="bold")
plt.xlabel("Unidade Federativa (UF)", fontsize=14)
plt.ylabel("")          # remove o título do eixo Y
plt.yticks([])          # remove os rótulos do eixo Y
plt.xticks(rotation=45) # evita sobreposição dos estados
plt.legend(title="Região do Brasil", fontsize=12, title_fontsize=13)

plt.show()
```

```python
#CONTRIBUIÇÃO DO COLEGA RODRIGO

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Carregar a base
df = pd.read_csv("acidentes2025.csv", encoding="ISO-8859-1", sep=";")

# Converter a coluna de data para datetime
df["data_inversa"] = pd.to_datetime(df["data_inversa"], format="%Y-%m-%d")

# Filtrar apenas o ano de 2025
df_2025 = df[df["data_inversa"].dt.year == 2025]

# Agrupar por mês
acidentes_por_mes = (
    df_2025.groupby(df_2025["data_inversa"].dt.month)
    .size()
    .reset_index(name="Total_Acidentes")
)

# Mapear número do mês para nome do mês em português
meses = {
    1: "Janeiro", 2: "Fevereiro", 3: "Março", 4: "Abril",
    5: "Maio", 6: "Junho", 7: "Julho", 8: "Agosto",
    9: "Setembro", 10: "Outubro", 11: "Novembro", 12: "Dezembro"
}
acidentes_por_mes["Mês"] = acidentes_por_mes["data_inversa"].map(meses)

# Configurar estilo
sns.set_theme(style="whitegrid")

# Criar gráfico de linha
plt.figure(figsize=(12, 6))
sns.lineplot(
    data=acidentes_por_mes,
    x="Mês",
    y="Total_Acidentes",
    marker="o",
    color="darkblue"
)

# Personalizar layout
plt.title("Acidentes registrados por mês (2025)", fontsize=18, weight="bold")
plt.xlabel("Mês", fontsize=14, weight="bold")
plt.ylabel("")          # remove o título do eixo Y
plt.yticks([])          # remove os rótulos do eixo Y
plt.xticks(rotation=45)

# Adicionar valores acima e à esquerda dos pontos
for i, row in acidentes_por_mes.iterrows():
    plt.text(
        i - 0.1, row["Total_Acidentes"] + 5,  # desloca à esquerda e acima
        str(row["Total_Acidentes"]),
        ha="right", va="bottom", fontsize=10, weight="bold", color="black"
    )

plt.show()

```

```python
#CONTRIBUIÇÃO FERNANDA

import pandas as pd
import matplotlib.pyplot as plt

# Leitura do CSV
df = pd.read_csv(
    'acidentes2025.csv',
    sep=';',
    encoding='ISO-8859-1'
)

# Limpeza básica
df.columns = df.columns.str.strip()

df['estado_fisico'] = (
    df['estado_fisico']
    .astype(str)
    .str.strip()
    .str.title()
)

# Ordem dos dias da semana
ordem_dias = [
    'segunda-feira', 'terça-feira', 'quarta-feira',
    'quinta-feira', 'sexta-feira', 'sábado', 'domingo'
]

df['dia_semana'] = pd.Categorical(
    df['dia_semana'],
    categories=ordem_dias,
    ordered=True
)

# Tabela de contingência
tabela = pd.crosstab(df['dia_semana'], df['estado_fisico'])

# Ordem REAL da gravidade (com Não Informado)
ordem_gravidade = [
    'Ileso',
    'Lesões Leves',
    'Lesões Graves',
    'Óbito',
    'Não Informado'
]

tabela = tabela[[c for c in ordem_gravidade if c in tabela.columns]]

# Converter para proporção
tabela_pct = tabela.div(tabela.sum(axis=1), axis=0)

# Cores semânticas
cores = [
    '#4CAF50',  # Ileso - verde
    '#FFC107',  # Lesões Leves - amarelo
    '#FF9800',  # Lesões Graves - laranja
    '#F44336',  # Óbito - vermelho
    '#9E9E9E'   # Não Informado - cinza
]

# Plot
ax = tabela_pct.plot(
    kind='bar',
    stacked=True,
    figsize=(10, 6),
    color=cores
)

plt.xlabel('Dia da semana')
plt.ylabel('Proporção')
plt.title('Proporção do estado físico das vítimas por dia da semana')
plt.legend(title='Estado físico')

# Adicionar porcentagens nas barras
for container in ax.containers:
    ax.bar_label(
        container,
        labels=[
            f'{v*100:.1f}%' if v >= 0.03 else ''  # evita poluição visual
            for v in container.datavalues
        ],
        label_type='center',
        fontsize=8
    )

plt.tight_layout()
plt.show()
```

```python
#CONTRIBUIÇÃO FERNANDA

import matplotlib.pyplot as plt

# Garantir padronização do estado_fisico
df['estado_fisico'] = (
    df['estado_fisico']
    .astype(str)
    .str.strip()
    .str.title()
)

# Contagem do estado físico
contagem = df['estado_fisico'].value_counts()

# Ordem lógica da gravidade (a mesma do gráfico de barras)
ordem_gravidade = [
    'Ileso',
    'Lesões Leves',
    'Lesões Graves',
    'Óbito',
    'Não Informado'
]

# Reordenar a contagem conforme a lógica de gravidade
contagem = contagem[[c for c in ordem_gravidade if c in contagem.index]]

# Cores semânticas (iguais às barras empilhadas)
cores = [
    '#4CAF50',  # Ileso - verde
    '#FFC107',  # Lesões Leves - amarelo
    '#FF9800',  # Lesões Graves - laranja
    '#F44336',  # Óbito - vermelho
    '#9E9E9E'   # Não Informado - cinza
]

# Plot do gráfico de pizza
plt.figure(figsize=(6, 6))
plt.pie(
    contagem.values,
    labels=contagem.index,
    autopct='%1.1f%%',
    startangle=90,
    colors=cores
)

plt.title('Distribuição do estado físico das vítimas')
plt.axis('equal')  # deixa o gráfico redondo
plt.tight_layout()
plt.show()
```

```python
#CONTRIBUIÇÃO FERNANDA

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# =========================
# 1. Leitura do CSV
# =========================
df = pd.read_csv(
    'acidentes2025.csv',
    sep=';',
    encoding='ISO-8859-1'
)

# =========================
# 2. Limpeza básica
# =========================
df.columns = df.columns.str.strip()

df['estado_fisico'] = (
    df['estado_fisico']
    .astype(str)
    .str.strip()
    .str.title()
)

df['tipo_veiculo'] = (
    df['tipo_veiculo']
    .astype(str)
    .str.strip()
    .str.title()
)

# =========================
# 3. Definir acidentes graves
# =========================
df_graves = df[df['estado_fisico'].isin(['Lesões Graves', 'Óbito'])]

# =========================
# 4. Cálculo por tipo de veículo
# =========================
total_por_tipo = df['tipo_veiculo'].value_counts()
graves_por_tipo = df_graves['tipo_veiculo'].value_counts()

comparacao = pd.DataFrame({
    'Total': total_por_tipo,
    'Graves': graves_por_tipo
}).fillna(0)

comparacao['Proporcao_Graves'] = comparacao['Graves'] / comparacao['Total']

# Remover tipos sem acidentes graves
comparacao = comparacao[comparacao['Proporcao_Graves'] > 0]

# Filtrar volume mínimo (opcional, mas recomendado)
comparacao = comparacao[comparacao['Total'] >= 50]

# Ordenar
comparacao = comparacao.sort_values('Proporcao_Graves', ascending=True)

# =========================
# 5. Gráfico com paleta neutra + %
# =========================
plt.figure(figsize=(9, 6))

cores = plt.cm.Greys(
    np.linspace(0.4, 0.85, len(comparacao))
)

bars = plt.barh(
    comparacao.index,
    comparacao['Proporcao_Graves'],
    color=cores
)

plt.xlabel('Proporção de acidentes graves')
plt.ylabel('Tipo de veículo')
plt.title('Proporção de acidentes graves por tipo de veículo')

# Adicionar percentual no fim de cada barra
for bar in bars:
    largura = bar.get_width()
    plt.text(
        largura + 0.005,
        bar.get_y() + bar.get_height() / 2,
        f'{largura*100:.1f}%',
        va='center',
        fontsize=9
    )

plt.tight_layout()
plt.show()
```
